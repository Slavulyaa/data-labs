```python
import pandas as pd
df_products= pd.read_pickle('webscrap_data_mais.pkl')
df_products
```

```python
df=df_products.copy()
```

```python
def clean_price(old_price):
    if isinstance(old_price, str):
        #print(old_price)
        price = old_price[:-2]
        price = price.replace('€', '.')
        return float(price)
    return old_price
```

```python
for col in df.columns:
    if "_price" in col:
        df[col] = df[col].apply(clean_price)
```

```python
df
```

```python
df.name = df.name.apply(lambda s: s.replace('ï', 'i'))
```

```python
#create a new column by culture type
def get_type(name):
    if 'Mais grain' in name:
        return 'Mais grain'
    if 'Mais mixte' in name:
        return 'Mais mixte'
    if 'Mais fourrage' in name:
        return 'Mais fourrager'
    
df['type']=df.name.apply(get_type)

#create a new column by brand
def get_brand(name):
    if 'DEKALB' in name:
        return 'DEKALB'
    if 'Limagrain' in name:
        return 'Limagrain'
    if 'Syngenta' in name:
        return 'Syngenta'
    else :
        return 'Others'
    
df['brand']=df.name.apply(get_brand)
df
```

```python
#change the place in the column
cols = df.columns.tolist()
column_to_move = "type"
new_position = 1

cols.insert(new_position, cols.pop(cols.index(column_to_move)))
df = df[cols]

column_to_move = "brand"
new_position = 0

cols.insert(new_position, cols.pop(cols.index(column_to_move)))
df = df[cols]
```

```python
df
```

```python
import math
products_offers = []

max_offers = max([int(c.split('_')[1]) for c in df.columns if '_price' in c])
print(max_offers)

for row in df.values:
    brand, type_, name,  = row[0], row[1], row[2]
    for ix_offer in range(max_offers):
        dpo = dict(brand=brand, name=name, type=type_)
        dpo['packaging'] = row[3+ix_offer*4]
        dpo['price'] = row[3+ix_offer*4+1]
        dpo['treatment'] = row[3+ix_offer*4+2]
        dpo['unit'] = row[3+ix_offer*4+3]
        if not math.isnan(dpo['price']):
            products_offers.append(dpo)

df_products_offers = pd.DataFrame.from_records(products_offers)
df_products_offers
```

### ANALYSIS

```python
s=round(df_products_offers.groupby('type')['price'].mean(),1)

df_mean_type = s.to_frame().T
df_mean_type
```

```python
sss=round(df_products_offers.groupby('treatment')['price'].mean(),1)
df_mean_treatment=sss.to_frame()
df_mean_treatment
```

```python
ss=round(df_products_offers.groupby('brand')['price'].mean(),1)
df_mean_brand= ss.to_frame().T
df_mean_brand
```
